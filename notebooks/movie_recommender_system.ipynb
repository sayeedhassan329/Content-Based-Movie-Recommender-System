3. Formating the data. 
From the features which are in the format of **string of list of dictionariers**, take the important list of values only.

```python
# to convert the string of list to only list:
import ast
```

```python
def convert(obj):
    list = []
    for i  in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# convert the keywords feature  like genres feature:
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
 # "For the 'cast' data, keep only the first 3 actor dictionaries since they correspond to the main starring roles."
# And from those dictionaries take out only the real names of actors and make list of those 3 names

def convert3(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

From the crew feature extracting only the director name , which we are interested in:
1. take out the dictionary in which name is director.
2. take out the value of name key from that dictionary.

```python
def fetch_director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew']  = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

Overview feature is a string, we need to convert it into list of words so that it can be concatenated with the others in the later stage.

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

Strip the spaces between the words of a single name

```python
for feature in ['genres', 'keywords', 'cast', 'crew']:
    movies[feature] = movies[feature].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

Now create a new feature named tags which will be the concatenaton of overview, genres, keywords, cast, crew

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

Create a new_df with only interested features

```python
new_df = movies[['movie_id', 'title', 'tags']].copy()
```

```python
new_df.head()
```

Make a single string out of the list in the tags featue:

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
# convert the tags feature in lowercase
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

## model training

Apply stemming to the tags feature to remove related words like loved, loving, loved = love etc

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

Use count_vectorizer of scikit learn. 
What it does:
1. Makes a bag of words from all the rows of the selected column and returns their frequency appearance.
2. Returns the top frequent words with 'max_features' after removing stop words with 'stop_words='english'.
3. Finally checks the appearance of each word in a row and returns their frequency, which makes the vector of that row.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# we can get the 5000 words which the cv object returned:
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

Now we need to find the similarity between the vectors, we will count the cosine distance (angle) between them.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

Make a function (recommender) which will return top n similar movies from the similarity matrix:
Note: This top n similar movies =(sorting) method will shuffle the movies, therefore we need to append index with movie so that where it goes in sorting, it goes with index intact.
 1. Get the index of the movie of which trying to return the similar movies from the new_df.
 2. Get the similarity array for that particular index from the similarity matrix. This similarity array is the similarity of all the movies with the particular index movie.
 3. Sort the array with every values index intact(tuple of index and similarity value) in reverse order to  get the top n values ( which is the index and the similarity value ).
 4. Take out the index values and match with the title feature of the new_df dataframe.

```python
def recommender(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]  # key is for sorting the tuple on second value and [1:6] is for getting top 5 movies, since 0 is itself.

    rec_list = []
    for i in movies_list:# to get the only index of the movies and match them with title and get movie names
        rec = new_df.iloc[i[0]].title
        rec_list.append(rec)
    return rec_list
```

```python
recommender('Batman Begins')
```

## Using Nearest Neighbors algorithm

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=6, metric='cosine')
model.fit(vectors)
```

```python
def recommend_movies(movie_title):
    # Find movie index
    movie_index = new_df[new_df['title'] == movie_title].index[0]

    movie_vector = vectors[movie_index].reshape(1, -1)
    # Get nearest neighbors using KNN
    distances, indices = model.kneighbors(movie_vector)
    
    # Return top 5 similar movies (excluding itself)
    similar_movies = new_df.iloc[indices[0][1:6]]['title']
    return similar_movies
```

## Evaluation:
### use diversity score metric:
    The ideal score depends on your specific goals:
    
        For discovery-focused systems (where you want to surprise users):
    
            Target: 0.1-0.3
    
            Example: "If you liked The Dark Knight, try this documentary about Batman's history"
    
        For precision-focused systems (where similarity is key):
    
            Target: 0.4-0.6
    
            Example: "If you liked The Dark Knight, try Batman Begins"
    
        Balanced systems:
    
            Target: 0.2-0.4

Real-World Examples:
| Platform | Strategy                          | Diversity Level       |
|----------|-----------------------------------|-----------------------|
| Netflix  | "Because you watched X" + "Trending" | Moderate (0.3–0.5)    |
| Spotify  | "Discover Weekly" playlists       | High (0.1–0.3)        |
| Amazon   | "Frequently bought together"      | Low (0.6–0.8)         |

```python

movie_indices = {title: idx for idx, title in enumerate(new_df['title'])}
```

```python
def calculate_diversity_bow(recommendations, count_vectorizer, movie_indices, bow_matrix):
    """
    Calculate recommendation diversity using Bag of Words vectors
    
    Parameters:
    - recommendations: List of recommended movie titles
    - count_vectorizer: Your fitted CountVectorizer instance
    - movie_indices: Dictionary mapping movie titles to matrix indices
    - bow_matrix: Your bag-of-words matrix from CountVectorizer
    
    Returns:
    - Diversity score (lower = more diverse)
    """
    # Get indices for recommended movies
    rec_indices = [movie_indices[title] for title in recommendations]
    
    # Extract BOW vectors for these movies
    rec_vectors = bow_matrix[rec_indices]
    
    # Calculate pairwise cosine similarities
    similarity_matrix = cosine_similarity(rec_vectors)
    
    # Zero out diagonal (self-similarity)
    np.fill_diagonal(similarity_matrix, 0)
    
    # Return average similarity between different movies
    return similarity_matrix.mean()

# Example usage:
diversity_scores = []
  # Your sample movies

# Assuming you have:
# - count_vect: Your fitted CountVectorizer
# - bow_matrix: The output from count_vect.fit_transform()
# - movie_indices: Dictionary like {'The Dark Knight': 0, 'Inception': 1, ...}

for movie in new_df['title'].sample(100):
    recs = recommender(movie)
    diversity = calculate_diversity_bow(recs, cv, movie_indices, vectors)
    diversity_scores.append(diversity)
    # print(f"Recommendations for {movie}:")
    # print(recs)
    # print(f"Diversity score: {diversity:.3f}\n")

print(f"\nAverage diversity across all samples: {np.mean(diversity_scores):.3f}")
print("(Lower values indicate more diverse recommendations)")
```

```python

def brute_force_nearest(title, all_vectors = vectors, n=5, metric='cosine'):
    query_vector = vectors[movie_indices[title]]
    query_2d = np.reshape(query_vector, (1, -1))
    distances = pairwise_distances(query_2d, all_vectors, metric=metric).flatten()
    nearest_indices = np.argsort(distances)[:6]  # For cosine similarity, lower = closer
    list = []
    for index in nearest_indices:
        list.append(movie_title[index])
    return list
```
